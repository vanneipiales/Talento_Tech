# **Ejercitación descubre el error**
 Qué tipo de error arroja el siguiente código, o no hay error?

** Respuesta:** El código no tiene errores evidentes para casos de uso normales con entrada numérica, pero carece de manejo de errores para entradas no válidas, lo que podría considerarse un error de diseño en un programa real.

```python
try:
    temperature = float(input('Ingresa la temperatura actual:'))

    if temperature > 0:
        print("Por encima de cero")
    elif temperature < 0:
        print("Por debajo de cero")
    else:
        print("Cero")

except ValueError:
    print("Error: Debes ingresar un número válido")
```

# **¿Qué sucede si ejecuto el código con un valor como 20 o 0?**

1. Probando con temperatura = 20

```python
temperature = float(input('Ingresa la temperatura actual:'))  # Usuario ingresa 20
if temperature > 0:
    print("Por encima de cero")  # Esta línea se ejecuta
elif temperature < 0:
    print("Por debajo de cero")  # Se omite
else:
    print("Cero")  # Se omite
```

2. Probando con temperatura = 0

```python
temperature = float(input('Ingresa la temperatura actual:'))  # Usuario ingresa 0
if temperature > 0:
    print("Por encima de cero")  # Se omite
elif temperature < 0:
    print("Por debajo de cero")  # Se omite
else:
    print("Cero")  # Esta línea se ejecuta
```
