# **Ejercitación recursividad**

 Ya conoce la función factorial y la sucesión de Fibonacci, incluso
 una implementación tradicional con funciones, su tarea es
 implementar dichas funciones utilizando recursividad.
 Ejercitación El juego del ahorcado

```python
#1. Factorial Recursivo

def factorial(n):
    # Caso base: 0! = 1 o 1! = 1
    if n == 0 or n == 1:
        return 1
    # Caso recursivo: n! = n * (n-1)!
    else:
        return n * factorial(n - 1)

print(factorial(5))  # Salida: 120
```

```python
#2. Sucesión de Fibonacci Recursivo

def fibonacci(n):
    # Casos base: fib(0) = 0, fib(1) = 1
    if n == 0:
        return 0
    elif n == 1:
        return 1
    # Caso recursivo: fib(n) = fib(n-1) + fib(n-2)
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)

print(fibonacci(6))  # Salida: 8
```

# **Ejercitación El juego del ahorcado**



```python
import random
import os

def limpiar_pantalla():
    os.system('cls' if os.name == 'nt' else 'clear')

def main():
    # Lista de palabras para el juego
    palabras = [
        "PYTHON", "PROGRAMACION", "AHORCADO", "COMPUTADORA", "TECLADO",
        "DESARROLLADOR", "ALGORITMO", "VARIABLE", "FUNCION", "DICCIONARIO",
        "LISTA", "TUPLA", "CONDICIONAL", "BUCLE", "RECURSIVIDAD"
    ]

    # Arte ASCII del ahorcado por etapas
    etapas_ahorcado = [
        """
          +---+
          |   |
              |
              |
              |
              |
        =========
        """,
        """
          +---+
          |   |
          O   |
              |
              |
              |
        =========
        """,
        """
          +---+
          |   |
          O   |
          |   |
              |
              |
        =========
        """,
        """
          +---+
          |   |
          O   |
         /|   |
              |
              |
        =========
        """,
        """
          +---+
          |   |
          O   |
         /|\\  |
              |
              |
        =========
        """,
        """
          +---+
          |   |
          O   |
         /|\\  |
         /    |
              |
        =========
        """,
        """
          +---+
          |   |
          O   |
         /|\\  |
         / \\  |
              |
        =========
        """
    ]

    # Función para obtener una palabra válida
    def obtener_palabra_valida(palabras):
        palabra = random.choice(palabras)
        while '-' in palabra or ' ' in palabra:
            palabra = random.choice(palabras)
        return palabra.upper()

    # Función para mostrar el estado del juego
    def mostrar_tablero(intentos, letras_adivinadas, letras_erroneas):
        print(etapas_ahorcado[intentos])
        print("\nLetras incorrectas:", ' '.join(letras_erroneas))
        print("\nPalabra:", ' '.join(letras_adivinadas))

    # Lógica principal del juego
    def jugar_ahorcado():
        palabra = obtener_palabra_valida(palabras)
        letras_palabra = set(palabra)
        letras_adivinadas = ['_'] * len(palabra)
        letras_erroneas = []
        intentos = 0
        abecedario = set('ABCDEFGHIJKLMNOPQRSTUVWXYZ')
        letras_usadas = set()

        limpiar_pantalla()
        print("¡Bienvenido al juego del ahorcado!")
        print(f"Adivina la palabra de {len(palabra)} letras\n")

        while len(letras_palabra) > 0 and intentos < len(etapas_ahorcado) - 1:
            mostrar_tablero(intentos, letras_adivinadas, letras_erroneas)

            letra = input("\nAdivina una letra: ").upper()

            if letra in abecedario - letras_usadas:
                letras_usadas.add(letra)
                if letra in letras_palabra:
                    letras_palabra.remove(letra)
                    for i, l in enumerate(palabra):
                        if l == letra:
                            letras_adivinadas[i] = letra
                else:
                    intentos += 1
                    letras_erroneas.append(letra)
            elif letra in letras_usadas:
                print("\nYa usaste esa letra. Intenta con otra.")
            else:
                print("\nEsa no es una letra válida.")

        limpiar_pantalla()
        mostrar_tablero(intentos, letras_adivinadas, letras_erroneas)

        if intentos == len(etapas_ahorcado) - 1:
            print(f"\n¡Perdiste! La palabra era: {palabra}")
        else:
            print("\n¡Felicidades! ¡Adivinaste la palabra!")

        print("\nGracias por jugar")

    # Iniciar el juego
    jugar_ahorcado()

if __name__ == "__main__":
    main()
```
